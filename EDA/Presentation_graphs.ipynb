### Presentacion

En este notebook los graficos para la presentacion del projecto en powerpont

```python
import numpy as np
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",40)
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("path_a_los_datos" , sep = "|")
```

#### target

```python
plt.figure(figsize=(7,5))
ax = sns.countplot(x="target",data=df,palette='RdBu_r')
```

#### temporales

```python
df[df['target']==0]['TENURE_CUSTOMER'].hist(bins=30,alpha=0.7, label= "target = 0")
df[df['target']==1]['TENURE_CUSTOMER'].hist(bins=30,color='darkred',alpha=0.7, label= "target = 1")

plt.xlabel('Antiguedad cliente (meses)')
plt.ylabel('Cantidad de clientes')
plt.title('TENURE_CUSTOMER')

plt.legend(loc=0)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2,figsize=(7,5),sharey=True,sharex= True)

axs[0] = (df[df['target']==0]['PREP_RECH_NDAYS_LASTRECH_12W'].hist(bins=30,alpha=0.7,ax=axs[0],label= "target = 0"))
axs[1] = df[df['target']==1]['PREP_RECH_NDAYS_LASTRECH_12W'].hist(bins=30,color='darkred',alpha=0.7, label= "target = 1")
plt.suptitle("PREP_RECH_NDAYS_LASTRECH_12W")

axs[0].set_ylabel('Cantidad de clientes')
axs[0].set_xlabel('Cant. dias desde la ultima recarga')
axs[1].set_xlabel('Cant. dias desde la ultima recarga')

axs[0].legend(loc=0)
axs[1].legend(loc=0)
```
